```python
import networkx as nx
import pandas as pd
```

```python
# Initialize the multiedge directed graph
G = nx.MultiDiGraph()

# Add edges from DataFrame
G.add_edges_from(filtered_nn_signaling[['from', 'to']].itertuples(index=False, name=None))
```

```python
G
```

```python
from collections import Counter
```

```python
def filter_multiedges(G):
    # Create a new graph to store filtered edges
    filtered_graph = nx.MultiDiGraph()

    # Get a count of all edge occurrences between node pairs
    edge_count = Counter((u, v) for u, v, _ in G.edges(keys=True))

    # Iterate over the edges and only add those that have 2 or more copies
    for (u, v), count in edge_count.items():
        if count >= 7:
            # Add all edges between u and v if they have 2 or more copies
            for key in G[u][v]:  # Get the keys of all multiedges
                filtered_graph.add_edge(u, v)

    return filtered_graph

# Filter the graph
filtered_G = filter_multiedges(G)
```

```python
combined_cellchatdb_nn_signaling = set(filtered_G.nodes)
```

```python
combined_c_n = { 'combined_c_n' : combined_cellchatdb_nn_signaling}
```

```python
with open('./combined_cellchatdb_nn_signaling.pickle', 'wb') as handle:
    pickle.dump(combined_c_n, handle, protocol=4)
```

## relax the thresholds

```python
def filter_multiedges_relax(G, thresh):
    # Create a new graph to store filtered edges
    filtered_graph = nx.MultiDiGraph()

    # Get a count of all edge occurrences between node pairs
    edge_count = Counter((u, v) for u, v, _ in G.edges(keys=True))

    # Iterate over the edges and only add those that have 2 or more copies
    for (u, v), count in edge_count.items():
        if count >= thresh:
            # Add all edges between u and v if they have 2 or more copies
            for key in G[u][v]:  # Get the keys of all multiedges
                filtered_graph.add_edge(u, v)

    return filtered_graph

# Filter the graph
filtered_G_5 = filter_multiedges_relax(G, 5)
filtered_G_6 = filter_multiedges_relax(G, 6)
```

```python
combined_cellchatdb_nn_signaling_5 = set(filtered_G_5)
combined_cellchatdb_nn_signaling_6 = set(filtered_G_6)
```
