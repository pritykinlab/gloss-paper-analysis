```python
# Initialize the multiedge directed graph
G = nx.MultiDiGraph()

# Add edges from DataFrame
G.add_edges_from(cd40_set[['from', 'to']].itertuples(index=False, name=None))
```

```python
from collections import Counter

def find_multiedges_out_from_node(G, node):
    # Get all outgoing edges from the specified node
    out_edges = G.out_edges(node)
    
    # Extract only the target nodes for these edges
    target_nodes = [target for _, target in out_edges]
    
    # Count occurrences of each target node
    target_counts = Counter(target_nodes)
    
    # Filter nodes with more than one edge from the specified node
    multiedge_nodes = {target: count for target, count in target_counts.items() if count > 1}
    
    return multiedge_nodes

# Example usage
multiedge_nodes = find_multiedges_from_node(G, 'Cd40')
```

```python
multiedge_nodes
```

```python
from collections import Counter

def find_multiedges_from_node_out(G, node):
    # Get all outgoing edges from the specified node
    out_edges = G.out_edges(node)
    
    # Extract only the target nodes for these edges
    target_nodes = [target for _, target in out_edges]
    
    # Count occurrences of each target node
    target_counts = Counter(target_nodes)
    
    # Filter nodes with more than one edge from the specified node
    multiedge_nodes = {target: count for target, count in target_counts.items() if count > 1}
    
    return multiedge_nodes

# Example usage
multiedge_nodes_out = find_multiedges_from_node_out(G, 'Cd40')
```

```python
from collections import Counter

def find_multiedges_from_node_in(G, node):
    # Get all outgoing edges from the specified node
    in_edges = G.in_edges(node)
    
    # Extract only the target nodes for these edges
    target_nodes = [source for source, _ in in_edges]
    
    # Count occurrences of each target node
    target_counts = Counter(target_nodes)
    
    # Filter nodes with more than one edge from the specified node
    multiedge_nodes = {target: count for target, count in target_counts.items() if count > 1}
    
    return multiedge_nodes

# Example usage
multiedge_nodes_in = find_multiedges_from_node_in(G, 'Cd40')
```

```python
cd40_filtered = {
    'cd40_f_in' : list(cd40_filtered_in),
    'cd40_f_out' : list(cd40_filtered_out),
    'cd40_f_union' : list(cd40_filtered_out.union(cd40_filtered_in)),
    'cd40_f_intersection' : list(cd40_filtered_out.intersection(cd40_filtered_in))
}
```
